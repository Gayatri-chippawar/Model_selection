## Model selection 

comparing the model with default hyperparameter value using cross validation 

```python
#list of models 
models=[LogisticRegression(max_iter=10000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
def cross_val_comparision():
    for model in models:
        cv_score=cross_val_score(model,X,Y,cv=5)
        mean_accuracy=sum(cv_score)/len(cv_score)
        print(f"the cross validation score of {model} is ={cv_score}")
        print(f"the accuracy score of the {model} is:{mean_accuracy*100:.2f}%")
        print("-------------------------------------------------------------")
```

```python
cross_val_comparision()
```

**Note-->** the Logistic regression has the highest accuracy among all the other three models

**Note2-->** the KNeighborclassifier has the lowest accurcy among all of them
